```python
import cv2
from ultralytics import YOLO

model = YOLO("/kaggle/working/runs/detect/train/weights/best.pt")

cap = cv2.VideoCapture("/kaggle/input/parking/busy_parking_lot.mp4")


fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter("output.mp4", fourcc, 20.0, 
                      (int(cap.get(3)), int(cap.get(4))))

while True:
    ret, frame = cap.read()
    if not ret:
        break

    results = model(frame)
    annotated = results[0].plot()

    empty, occupied = 0, 0
    for box in results[0].boxes:
        cls = int(box.cls[0])
        if cls == 0:
            empty += 1
        elif cls == 1:
            occupied += 1

    cv2.putText(annotated, f"Empty: {empty} | Occupied: {occupied}",
                (30, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 255), 2)


    out.write(annotated)

cap.release()
out.release()
```
